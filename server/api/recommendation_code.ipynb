```python
def recommend_song(last_word):
    df1=df.copy()
    df1['Lyric'] = df1['Lyric'].str.split().str[:5]
    #print(df1['Lyric'])
    rec_song=[]
    for i in range(len(df1)):
        k=df1['Lyric'][i]
        #print(k)
        for j in range(len(k)):
            if k[j].lower()==last_word.lower():
        # if last_word in df1[i]['Lyric']:
                rec_song.append([i,df1['SName'][i],df1['Genres'][i].split('; ')])
    return rec_song
#     print(rec_song)

print(recommend_song('joker'))
    
```

```python
def recommend_song_genre(word,genre):
    rec_song=recommend_song(word)
    genre=[i.lower() for i in genre]
    rec_song=[[k,j,[y.lower() for y in i]] for k,j,i in rec_song]
    ans=[]
    for i in range(len(rec_song)):
        
        lst1=list(set(genre)&set(rec_song[i][2]))
        if len(lst1)>0:
            ans.append([len(lst1),rec_song[i][0]])
    ans.sort(reverse=True)
    lst=[]
    for k,s in ans:
        lst.append(s)
    #print(lst)
    return lst
    
recommend_song_genre('love',['poP'])
    
```

```python

import random
def get_user(genre):
    user=[]
    for i in range(50):
        r1= random.randint(1, 4)
        useri=[i]
        lst=random.choices(genre,k=r1)
        lst=",".join(lst)
        useri.append(lst)
        user.append(useri)
    return user
user=get_user(genre)
print(user)
tocsv=pd.DataFrame(user,columns=['userId','genres'])
tocsv.to_csv("users.csv",index=False)
```
